```python
import cv2
from matplotlib import pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('sanlokhan.JPG')
```

```python
print('Original dimensions:',img.shape)
```

```python
width = int(img.shape[1]*0.2)
height = int(img.shape[0]*0.2)
dim = (width,height)
```

```python
resized = cv2.resize(img,dim,interpolation=cv2.INTER_AREA)
```

```python
print('Resized dimensions:',resized.shape)
```

```python
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY))
```

```python
resizedGray = cv2.cvtColor(resized,cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(resizedGray,1.1,4)
```

```python
for (x ,y ,w ,h) in faces:
      cv2.rectangle(resized,(x , y),( x+w, y+h),(255,0,0),3)
```

```python
cv2.imshow('img',resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
